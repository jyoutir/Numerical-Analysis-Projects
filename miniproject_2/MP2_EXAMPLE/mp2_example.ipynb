```python
#
# FUNCTION DEFINITIONS
#

def geometric(r,N):
    S = 0.0
    for n in range(1,N+1):
        S += r**n
    return S

def limit(r):
    return r/(1.0-r)

def generatelist(r,eps):
    difflist=[] 
    diff = 10.0 
    n = 1       
    while diff > eps:
        diff = limit(r)-geometric(r,n) 
        difflist.append(diff)
        n += 1
    return difflist

#
# MAIN SCRIPT
#

import matplotlib.pyplot as pl
eps = 1e-14
diff05=generatelist(0.5,eps)
diff07=generatelist(0.7,eps)
pl.semilogy(range(1,len(diff05)+1),diff05,':o',label='r=0.5')
pl.semilogy(range(1,len(diff07)+1),diff07,':x',label='r=0.7')
pl.xlabel('Terms in the sum')
pl.ylabel('Difference with limiting value')
pl.legend()
pl.savefig('fig.pdf')
```
